```python
import os
import pydicom
import pandas as pd
from collections import defaultdict

def extract_dicom_metadata(dicom_folder):
    metadata_list = []
    series_dict = defaultdict(int)  

    for root, _, files in os.walk(dicom_folder):
        for file in files:
            dicom_path = os.path.join(root, file)
            
            if file.lower().endswith(".dcm"):  
                try:
                    dicom_data = pydicom.dcmread(dicom_path)
                   
                    patient_id = dicom_data.get("PatientID", "Unknown")
                    study_uid = dicom_data.get("StudyInstanceUID", "Unknown")
                    series_uid = dicom_data.get("SeriesInstanceUID", f"Unknown_{file}")
                    slice_thickness = dicom_data.get("SliceThickness", "Unknown")
                    pixel_spacing = dicom_data.get("PixelSpacing", "Unknown")
                    study_date = dicom_data.get("StudyDate", "Unknown")
                    acquisition_date = dicom_data.get("AcquisitionDate", "Unknown")

                    series_dict[series_uid] += 1

                    metadata_list.append({
                        "PatientID": patient_id,
                        "StudyInstanceUID": study_uid,
                        "SeriesInstanceUID": series_uid,
                        "SliceThickness": slice_thickness,
                        "PixelSpacing": pixel_spacing,
                        "StudyDate": study_date,
                        "AcquisitionDate": acquisition_date,
                        "DICOM_File": dicom_path
                    })
                
                except Exception as e:
                    print(f"Error reading {dicom_path}: {e}")

    metadata_df = pd.DataFrame(metadata_list)

    if not metadata_df.empty and "SeriesInstanceUID" in metadata_df:
        metadata_df["NumSlicesPerSeries"] = metadata_df["SeriesInstanceUID"].map(series_dict)

    return metadata_df

if __name__ == "__main__":
    dicom_folder = input("Enter the path to the DICOM folder: ").strip()
    
    if os.path.exists(dicom_folder) and os.path.isdir(dicom_folder):
        metadata_df = extract_dicom_metadata(dicom_folder)
        output_file = "dicom_metadata.csv"
        metadata_df.to_csv(output_file, index=False)
        print(f"Metadata extraction complete. Saved as {output_file}.")
    else:
        print("Invalid folder path! Please provide a valid DICOM folder.")
```

```python
CSV_FILE = os.path.abspath("dicom_metadata.csv")  
import sqlite3

DB_FILE = "dicom_metadata.db"
conn = sqlite3.connect(DB_FILE)
cursor = conn.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS dicom_metadata (
        PatientID TEXT,
        StudyInstanceUID TEXT,
        SeriesInstanceUID TEXT,
        SliceThickness TEXT,
        PixelSpacing TEXT,
        StudyDate TEXT,
        AcquisitionDate TEXT,
        DICOM_File TEXT,
        NumSlicesPerSeries INTEGER
    )
""")
if os.path.exists(CSV_FILE):
    df = pd.read_csv(CSV_FILE)
    df.to_sql("dicom_metadata", conn, if_exists="replace", index=False)
    print(f" Data inserted into {DB_FILE}")
else:
    print("CSV file not found! Make sure metadata extraction is complete.")
conn.close()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

CSV_FILE = os.path.abspath("dicom_metadata.csv")  # Update path if needed
df = pd.read_csv(CSV_FILE)

# Total number of studies
total_studies = df["StudyInstanceUID"].nunique()
print(f"Total number of studies: {total_studies}")

# Total slices across all scans
total_slices = df["NumSlicesPerSeries"].sum()
print(f"Total slices across all scans: {total_slices}")

# Average number of slices per study
average_slices_per_study = df.groupby("StudyInstanceUID")["NumSlicesPerSeries"].sum().mean()
print(f"Average slices per study: {average_slices_per_study:.2f}")

# Distribution of Slice Thickness
df["SliceThickness"] = pd.to_numeric(df["SliceThickness"], errors="coerce") 
df["PixelSpacing"] = df["PixelSpacing"].astype(str)

# Visualization: Histogram of Slice Thikness
plt.figure(figsize=(8, 5))
sns.histplot(df["SliceThickness"].dropna(), bins=20, kde=True, color="skyblue")
plt.xlabel("Slice Thickness (mm)")
plt.ylabel("Frequency")
plt.title("Distribution of Slice Thickness")
plt.grid(True)
plt.show()

# Visualization: Bar Chart of Slices per Series
plt.figure(figsize=(12, 6))
sns.barplot(x=df["SeriesInstanceUID"], y=df["NumSlicesPerSeries"], palette="coolwarm")
plt.xticks(rotation=90)
plt.xlabel("Series Instance UID")
plt.ylabel("Number of Slices")
plt.title("Number of Slices per Series")
plt.grid(axis="y")
plt.show()
```
